### **Internet Penetracion (DataSet_14)**
La Tasa de Penetración de Internet corresponde al porcentaje de la población total de un determinado país o región que utiliza Internet, en nuestro caso el procentaje es respecto a cada 100 hogares y cada 100 habitantes

```python
DataFrame_14 = pd.read_csv('datasets\DataSet_14.csv',decimal=',',thousands='.')
DataFrame_14.head(5)
```

```python
DataFrame_14['Relación'] = DataFrame_14['Accesos por cada 100 hab']/DataFrame_14['Accesos por cada 100 hogares'] 
DataFrame_14.head()
```

```python
DataFrame_14.info()
```

```python
DataFrame_14 = DataFrame_14.sort_values('Año')
fig, axes = plt.subplots(3,1 , figsize=(10, 15),sharex=True)
fig.suptitle('Acceso a Internet en el Tiempo')

# Accesos por cada 100 hogares
sns.lineplot(ax=axes[0],data=DataFrame_14, x='Periodo', y='Accesos por cada 100 hogares')
plt.xticks(rotation=90)
axes[0].set_title('Accesos por cada 100 hogares')

# Accesos por cada 100 hab
sns.lineplot(ax=axes[1],data=DataFrame_14, x='Periodo', y='Accesos por cada 100 hab')
plt.xticks(rotation=90)
axes[1].set_title('Accesos por cada 100 hab')

#Relación 'Accesos por cada 100 hab'/'Accesos por cada 100 hogares'
sns.scatterplot(ax=axes[2],data=DataFrame_14, x='Periodo', y='Relación')
plt.xticks(rotation=90)
axes[2].set_title('Relacion Acceso Habitantes/Hogares')
```

Se observa un incremento en el porcentaje de la población total con acceso a Internet por cada 100 hogares y cada 100 habitantes con una relación de aproximadamente 1/3 entre estas dos.

### **Internet Penetracion (DataSet_15)**

```python
DataFrame_15 = pd.read_csv('datasets\DataSet_15.csv',decimal=',',thousands='.')
DataFrame_15.head(5)
```

```python
DataFrame_15.info()
```

```python
DataFrame_15_PT=DataFrame_15.pivot_table(index='Provincia', columns='Año', values='Accesos por cada 100 hogares', aggfunc=np.average)
DataFrame_15_PT
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc = {'figure.figsize':(15,8)})
ax = sns.scatterplot(data=DataFrame_15_PT)
ax.collections[0].set_sizes([100]) 
ax = sns.color_palette("Set2")
plt.xticks(rotation=90)
plt.title('Promedio por Año en Acceso a Internet por Cada 100 Hogares')
```

En general se presentan más accesos a internet por provincia, donde Capital Federal apenas se a aumentado en sus accesos en comparación con el resto de capitales en las que se aprecia un rango considerable de accesos con los años
