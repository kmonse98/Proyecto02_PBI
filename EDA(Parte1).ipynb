A continuación deiniremos una función **VelocidadInternet_Provincia(provincia:str)** en la cual veremos con más detalle el comportamiento de la vlocidad de bajada para cada provincia.Los argumentos d la función son las provincias.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
DataFrame_5['Año']=DataFrame_5['Año'].astype('str')
DataFrame_5['Trimestre']=DataFrame_5['Trimestre'].astype('str')
DataFrame_5['Año/Trimestre'] = DataFrame_5.Año.str.cat(DataFrame_5.Trimestre, sep='-')
DataFrame_5=DataFrame_5.sort_values('Año')
```

```python
def VelocidadInternet_Provincia(provincia:str):
        print(provincia)
        df = DataFrame_5[DataFrame_5['Provincia']==provincia]
        sns.lineplot(data = df, x = "Año/Trimestre", y = "Mbps (Media de bajada)",color='red')
        plt.title('Velocidad Bajada de Internet (Mbps)')
        plt.xticks(rotation=90)
        #plt.title(provincia)
```

```python
VelocidadInternet_Provincia('Buenos Aires')
```

En la gráfica para **Buenos Aires** podemos apreciar que entre subidas y bajadas en la velocidad de internet se tienen una crecimeinto general.

### **Conectividad al servicio de Internet (DataSet_6)**

```python
import pandas as pd
import numpy as np
DataFrame_6 = pd.read_csv('datasets\DataSet_6.csv')
DataFrame_6.head(5)
```

```python
DataFrame_6.info()
```

```python
DataFrame_6_Poblacion = DataFrame_6.loc[:,['Provincia','Poblacion']]
DataFrame_6_Poblacion_Agrupado = DataFrame_6_Poblacion.groupby('Provincia').sum()
DataFrame_6_Poblacion_Agrupado.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.title('Población por Provincia')
sns.barplot(x=DataFrame_6_Poblacion_Agrupado.index,y=DataFrame_6_Poblacion_Agrupado['Poblacion'])
plt.xticks(rotation=90)
```

La anterior gráfica muestra un estimado de la cantidad de población por provincia, se observa que **BUENOS AIRES** es la provincia con mayor población, siguindole en cantidad: **CABA**, **CORDOBA** y **SANTA FE**

```python
''' Remplazamos los 'SI' por 1 y los '--' por 0 '''
for nombre in['ADSL','CABLEMODEM','DIALUP','FIBRAOPTICA','SATELITAL','WIRELESS','TELEFONIAFIJA','3G','4G']:
    DataFrame_6[nombre].replace(['--','SI'],[0,1],inplace=True)

DataFrame_6.head(5)
```

```python
DataFrame_6_Tecnología=DataFrame_6.loc[:,['Provincia', 'ADSL', 'CABLEMODEM',
       'DIALUP', 'FIBRAOPTICA', 'SATELITAL', 'WIRELESS', 'TELEFONIAFIJA', '3G',
       '4G']]
DataFrame_6_Tecnología.head()
```

Agrupamos por provincia bajo la suma de las respectivas tecnologías para obtener el número de localidades que tienen accesos a esa tecnología para cada provincia

```python
DataFrame_6_Agrupado = DataFrame_6_Tecnología.groupby(by=['Provincia']).sum()
DataFrame_6_Agrupado.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(9,1 , figsize=(10, 20),sharex=True)
fig.suptitle('Cantidad de Localidades (Acceso por Tecnología)')

sns.barplot(ax=axes[0],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['ADSL'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[1],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['CABLEMODEM'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[2],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['DIALUP'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[3],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['FIBRAOPTICA'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[4],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['SATELITAL'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[5],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['WIRELESS'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[6],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['TELEFONIAFIJA'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[7],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['3G'])
plt.xticks(rotation=90)

sns.barplot(ax=axes[8],x=DataFrame_6_Agrupado.index, y=DataFrame_6_Agrupado['4G'])
plt.xticks(rotation=90)
```

De las anteriores gráficas podemos observar que **Buenos Aires**, **Santa Fe** y **Cordoba** presentan un mayor acceso en sus localidades a todas las teconologías, y que en general hay mayor acceso de localidades a la red **SATELITAL**

### **Internet Accesos por Tecnología (DataSet_7)**

```python
import pandas as pd
import numpy as np
DataFrame_7 = pd.read_csv('datasets\DataSet_7.csv',thousands='.',decimal=',')
DataFrame_7.head(5)
```

```python
DataFrame_7.info()
```

```python
DataFrame_7.iloc[840] #La última columna no es necesaria
```

```python
# Eliminamos la última fila
DataFrame_7.drop([840],axis=0,inplace=True) 
DataFrame_7.tail(3)
```

```python
DataFrame_7[['anio', 'eliminar1']] = DataFrame_7['Año'].str.split(' ', expand=True)
DataFrame_7[['Trim', 'eliminar2']] = DataFrame_7['Trimestre'].str.split(' ', expand=True)
```

```python
print(DataFrame_7['eliminar1'].unique())
print(DataFrame_7['eliminar2'].unique())
```

```python
DataFrame_7.drop(['eliminar1','eliminar2','Año','Trimestre'],axis=1,inplace=True)
DataFrame_7.info()
```

```python
DataFrame_7['anio'] = DataFrame_7['anio'].astype('int')
DataFrame_7['Trim'] = DataFrame_7['Trim'].astype('int')
```

```python
DataFrame_7.info()
```

Hacemos una agrupación por **año** y **Provincia** bajo el máximo valor

```python
DataFrame_7_GroupBy=DataFrame_7.groupby(by=['anio','Provincia']).max()
DataFrame_7_GroupBy
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(4,1 , figsize=(10, 15),sharex=True)
fig.suptitle('Internet Acceso por Tecnología')
sns.color_palette("husl", 9)

# Accesos por cada 100 hogares
sns.scatterplot(ax=axes[0],data=DataFrame_7_GroupBy, x='Provincia', y='ADSL',hue='anio')
plt.yscale('log')
plt.xticks(rotation=90)
axes[0].set_title('Gráfica 1')

# Accesos por cada 100 hab
sns.scatterplot(ax=axes[1],data=DataFrame_7_GroupBy, x='Provincia', y='Cablemodem',hue='anio')
plt.xticks(rotation=90)
plt.yscale('log')
axes[1].set_title('Gráfica 2')

# Accesos por cada 100 hab
sns.scatterplot(ax=axes[2],data=DataFrame_7_GroupBy, x='Provincia', y='Fibra óptica',hue='anio')
plt.xticks(rotation=90)
plt.yscale('log')
axes[2].set_title('Gráfica 3')

# Accesos por cada 100 hab
sns.scatterplot(ax=axes[3],data=DataFrame_7_GroupBy, x='Provincia', y='Wireless',hue='anio')
plt.xticks(rotation=90)
plt.yscale('log')
axes[3].set_title('Gráfica 4')
```

**Nota:** En la gráfica referente a **ADSL** podemos ver un claro decremento con los años en el acceso a esta tecnología. En cambio en **Cablemodem**, **Wireless** y **Fibra Optica** si hay un incremento en sus accesos siendo **Wireless** la tennología con un incremento más significativo a lo largo de los años
